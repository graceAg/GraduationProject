```python
train=[]
train_mfcc = []

train_y = []
train_mfcc_y = []


pad1d = lambda a, i: a[0:i] if a.shape[0]>i else np.hstack((a, np.zeros(i-a.shape[0])))


for a in['1', '2', '3', '4', '5']:
    audio_path = "data\\bad\\"+a+".wav"

    w, sr = librosa.load(audio_path)
    mfcc = librosa.feature.mfcc(w)

    paddedW = pad1d(w, 30000)
    #shape를 통일하는 과정 같다.
    train.append(paddedW)
    train_y.append(True)
    
for a in['1', '2', '3', '4', '5']:
    audio_path = "data\\good\\"+a+".wav"

    w, sr = librosa.load(audio_path)
    mfcc = librosa.feature.mfcc(w)

    paddedW = pad1d(w, 30000)
    #shape를 통일하는 과정 같다.
    train.append(paddedW)
    train_y.append(False)
#불리언으로 해도 되나? 데이터가 늘어날때 자동으로 입력하는 방법을 찾기
train_mfcc.append(w)
```

```python
ip = Input(shape=(train[0].shape))
hidden = Dense(128, activation='relu')(ip)
op = Dense(1, activation='sigmoid')(hidden)
model = Model(ip, op)
# input = 변수, output = 변수 식으로 하면 오류가 난다.

model.summary()
```

```python
train = np.asarray(train)
train_y = np.asarray(train_y) #이거 해야 오류안뜸


model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
model.fit(train, train_y)
```
